*Convert miliseconds to seconds* 

```python
# Some string values in duration that prevent conversion
df_subset['duration_ms'].loc[91952:91970]  # First sequence of missing values
```

```python
# Convert these values to NaN values 
df_subset['duration_ms'] = pd.to_numeric(df['duration_ms'], errors='coerce')
df_subset['duration_ms'].isnull().sum()
```

```python
# Possibility of takeing average numeric value for the 19 missing values
df_subset['duration_ms'].fillna(df_subset['duration_ms'].mean(), inplace=True)
df_subset['duration_ms'].isnull().sum()
```

```python
# Convert ms to min 
df_subset['duration_min'] = df_subset['duration_ms']/60000
df_subset.drop(columns=['duration_ms'], inplace=True)
df_subset.head(10)
```

```python
#Finding missing values on the dataset
df_subset.isnull().sum()
```

```python
# Found again missing values in music features 
df_subset[df.eq('n\x07').any(axis=1)]
df_subset.replace({'n\x07': np.NaN}, inplace=True)
```

```python
# Change datatypes of some colums
df_subset[['danceability', 'energy','loudness', 'speechiness', 
   'acoustics', 'instrumentalness', 'liveliness', 'valence', 'tempo']] = df_subset[['danceability', 'energy', 'loudness', 'speechiness',
                                                                                               'acoustics', 'instrumentalness', 'liveliness', 'valence', 'tempo' ]].apply(pd.to_numeric)
# Type conversion for the integer columns doesnt seem to work 
df_subset = df_subset.astype({'mode': int, 'key': int, 'time_signature': int}, errors='ignore')  # Cant convert NaN values to int, so ignore them
df_subset.info()
```
